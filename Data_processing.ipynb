```python
import os
from glob import glob

import pandas as pd
import numpy as np
```

```python
data_dir = '/gdrive/MyDrive/0. 인공지능 학습/아이펠 대전 3기/5. 해커톤/1차 해커톤/data'
save_data_dir = '/gdrive/MyDrive/0. 인공지능 학습/아이펠 대전 3기/5. 해커톤/1차 해커톤/processed_data'

# 아마존
amazon_credits_path = os.path.join(data_dir, "아마존", "credits.csv")
amazon_titles_path = os.path.join(data_dir, "아마존", "titles.csv")
# 넷플릭스
netflix_credits_path = os.path.join(data_dir, "넷플릭스", "credits.csv")
netflix_titles_path = os.path.join(data_dir, "넷플릭스", "titles.csv")
# 디즈니
disney_credits_path = os.path.join(data_dir, "디즈니", "credits.csv")
disney_titles_path = os.path.join(data_dir, "디즈니", "titles.csv")
# 파라마운트
paramount_credits_path = os.path.join(data_dir, "파라마운트", "credits.csv")
paramount_titles_path = os.path.join(data_dir, "파라마운트", "titles.csv")
```

```python
credits_paths = [amazon_credits_path, netflix_credits_path, disney_credits_path, paramount_credits_path]
credits_save_file_names = ["amazon_credits.csv", "netflix_credits.csv", "disney_credits.csv", "paramount_credits.csv"]

titles_paths = [amazon_titles_path, netflix_titles_path, disney_titles_path, paramount_titles_path]
titles_save_file_names = ["amazon_titles.csv", "netflix_titles.csv", "disney_titles.csv", "paramount_titles.csv"]
```

# credits 파일
- 중복 제거
- name column 제거

```python
def credits_process(credits_paths, save_data_dir, file_names):
    for i, path in enumerate(credits_paths):
        credits = pd.read_csv(path)
        # 중복 제거
        credits.drop_duplicates()

        # name column 제거
        del credits["name"]

        # csv 저장
        save_data_path = os.path.join(save_data_dir, file_names[i])
        print("저장 경로 :", save_data_path)
        credits.to_csv(save_data_path, index=False)
```

```python
credits_process(credits_paths, save_data_dir, credits_save_file_names)
```

# titles 파일
- 중복 제거
- IMDB ID column 제거

```python
def titles_process(titles_paths, save_data_dir, file_names):
    for i, path in enumerate(titles_paths):
        titles = pd.read_csv(path)
        # 중복 제거
        titles.drop_duplicates()

        # imdb_id column 제거
        del titles["imdb_id"]

        # csv 저장
        save_data_path = os.path.join(save_data_dir, file_names[i])
        print("저장 경로 :", save_data_path)
        titles.to_csv(save_data_path, index=False)
```

```python
titles_process(titles_paths, save_data_dir, titles_save_file_names)
```
